# Treasurey Arbitrage

## 1. Treasury Arbitrage

### 1.1
- I would long the older bond while shorting the newer bond, since both bonds have the same maturity date and different YTM, there exists a arbitrage opportunity in capturing the difference in YTM

### 1.2
- I would borrow funds from the dealer to enter the long position alongside collatoral for the loan. I would roll over the loan for a repo agreement. in terms of the short position, I would find ask the dealer to another investor who ownes the short position bonds and are willing to lend them to me. I would sell the bonds for cash, and use it as collatoral for the loan.

### 1.3
The risks for this trade include when the yield spread increases and I now have to put on more collatoral or close the trade, and credit risk.

### 1.4

```python
short_bond_price = 87
long_bond_price = 113
years_to_maturity = 8.49  # Changed from 5.5

short_bond_yield = (100 * 0.0162 + (100 - short_bond_price) / years_to_maturity) / ((100 + short_bond_price) / 2)
long_bond_yield = (100 * 0.0612 + (100 - long_bond_price) / years_to_maturity) / ((100 + long_bond_price) / 2) 

print(f"Short-term bond yield: {short_bond_yield:.4f} or {short_bond_yield*100:.2f}%")
print(f"Long-term bond yield: {long_bond_yield:.4f} or {long_bond_yield*100:.2f}%")
```

### 1.5

```python
def price_from_ytm(ytm, coupon_rate, face_value, years):
    # Convert to semi-annual parameters
    semi_annual_ytm = ytm / 2  # Divide annual YTM by 2
    semi_annual_coupon = (face_value * coupon_rate) / 2  # Divide annual coupon by 2
    num_periods = int(years * 2)  # Multiply years by 2 to get semi-annual periods
    
    price = 0
    
    # Sum all semi-annual coupon payments
    for t in range(1, num_periods + 1):
        price += semi_annual_coupon / (1 + semi_annual_ytm) ** t
    
    # Add principal
    price += face_value / (1 + semi_annual_ytm) ** num_periods
    
    return price

# Calculate:
short_bond_price = price_from_ytm(0.0465, 0.0162, 100, 5.0)
long_bond_price = price_from_ytm(0.0470, 0.0612, 100, 5.0)

print(f"Calculated short-term bond price: {short_bond_price:.2f}")
print(f"Calculated long-term bond price: {long_bond_price:.2f}")
```
